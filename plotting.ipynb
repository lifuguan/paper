```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
print((0.7274 - 0.6893) / 0.6893 * 100)
print((0.7414 - 0.6893) / 0.6893 * 100)
print((0.7534 - 0.6893) / 0.6893 * 100)
print((0.7598 - 0.6893) / 0.6893 * 100)
print((0.7623 - 0.6893) / 0.6893 * 100)
```

```python
with plt.style.context(['ieee', 'science']):
    dice_means, dice_std = (0.7977, 0.8314, 0.8448, 0.8545, 0.8525, 0.8588), (0.01, 0.01, 0.01, 0.01, 0.01, 0.01)
    iou_means, iou_std = (0.6893, 0.7274, 0.7414, 0.7534, 0.7598, 0.7623), (0.01, 0.01, 0.01, 0.01, 0.01, 0.01)
    sen_means, sen_std = (0.8115, 0.9685, 0.9714, 0.9669, 0.9744, 0.9296), (0.01, 0.01, 0.01, 0.01, 0.01, 0.01)
    hd_means, hd_std = (5.5049, 5.5861, 5.0647, 4.9601, 4.9210, 4.6224), (0.01, 0.01, 0.01, 0.01, 0.01, 0.01)
    improve = (0, 5.52, 7.55, 9.29, 10.22, 10.59)
    names = ['UNet', 'ResUNet', 'ResUNet\n+MRC\n+RMP', 'ResUNet\n+SAR', 'ResUNet\n+SAR\n+MRC\n+RMP', 'ResUNet\n+SAR\n+MRC\n+RMP\n+Loss']
    ind = np.arange(len(dice_means))  # the x locations for the groups
    width = 0.2  # the width of the bars
    fig, ax = plt.subplots(figsize=(6, 3))
    ax.set_ylim(0.6, 1)
    rects1 = ax.bar(ind - width * 1.5,iou_means, width,  label="Iou", color="silver")
    rects2 = ax.bar(ind - width * 0.5,dice_means, width, label="F1 score", color="skyblue")
    rects3 = ax.bar(ind + width * 0.5,dice_means, width, label='Dice', color="c")
    rects4 = ax.bar(ind + width * 1.5,sen_means, width, label='Sensitivity', color="darkcyan")
    ax.set_xticks(ind)
    ax.set_xticklabels(names)
    ax.set_ylabel('Metrics')
    ax.set_title('(1) Metric results of COVID19 segmentation')
    ax.grid(b=True, axis='y')

    ax2 = ax.twinx()
    rects5 = ax2.plot(improve,marker='s', label="Improvement", color="orange")
    ax2.set_xticks(ind)
    ax2.set_xticklabels(names)
    ax2.set_ylabel('IOU Improvement(\%)')
    fig.legend(loc="upper left", bbox_to_anchor=(0, 1), bbox_transform=ax.transAxes) 
    plt.show()
    fig.savefig("figure/cell_comparison.pdf")
```
